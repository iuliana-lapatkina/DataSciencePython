```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
t = np.linspace(0, 10, 51, endpoint=True)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t,f, color='green')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.title('График f(t)', fontsize=16, fontweight='bold')
plt.xlabel('Значения t', fontsize=12, color='#3d3d3d')
plt.ylabel('Значения f', fontsize=12, color='#3d3d3d')
plt.show()
```
